```python
# Test streaming access
import datasets
from datasets import load_dataset
import json

# Start with evaluation subset (smaller, faster)
print("Loading Egocentric-10K-Evaluation...")
ds = load_dataset(
    "builddotai/Egocentric-10K",
    "Egocentric-10K-Evaluation",
    streaming=True,
    split="train", 
    download_config=datasets.DownloadConfig(
        max_retries=5 # Increase from default 10s to 120s
    )
)
```

```python
# Examine first 5 samples
for i, sample in enumerate(ds.take(5)):
    print(f"\n=== Sample {i} ===")
    print(f"Keys: {sample.keys()}")

    # Look at metadata
    if "json" in sample: 
        metadata = json.loads(sample["json"])
        print(f"Factory: {metadata.get('factory_id')}")
        print(f"Worker: {metadata.get('worker_id')}")
        print(f"Duration: {metadata.get('duration')}s")
        print(f"Resolution: {metadata.get('width')}x{metadata.get('height')}")
        print(f"FPS: {metadata.get('fps')}")
    
    # Look at video
    if 'mp4' in sample: 
        print(f"Video size: {len(sample['mp4'])} bytes")

    if i >= 4: 
        break
```
